```python
import pandas as pd
import requests

# Constants
TMDB_API_KEY = 'b3ba374a1d289fb7620e885322528803'
SEARCH_URL = "https://api.themoviedb.org/3/search/movie"
MOVIE_DETAILS_URL = "https://api.themoviedb.org/3/movie/{}?append_to_response=credits"

# Read the CSV
df = pd.read_csv('C:/Users/14435/Downloads/tmdb_5000_movies_with_director (1).csv')

# Function to get director name using TMDB API
def get_director(movie_name):
    search_payload = {
        "api_key": TMDB_API_KEY,
        "query": movie_name
    }
    
    response = requests.get(SEARCH_URL, params=search_payload)
    data = response.json()
    if 'results' in data and data['results']:
        movie_id = data['results'][0]['id']
        details_response = requests.get(MOVIE_DETAILS_URL.format(movie_id), params={"api_key": TMDB_API_KEY})
        details = details_response.json()
        if 'credits' in details and 'crew' in details['credits']:
            for crew_member in details['credits']['crew']:
                if crew_member['job'] == 'Director':
                    return crew_member['name']
    return None

# Fill missing director names
for idx, row in df.iterrows():
    if pd.isna(row['Director']):
        director_name = get_director(row['original_title'])
        if director_name:
            df.at[idx, 'Director'] = director_name

# Save the updated dataframe to CSV
df.to_csv('C:/Users/14435/Downloads/tmdb_5000_movies_with_director_UPDATED3.csv', index=False)
```
